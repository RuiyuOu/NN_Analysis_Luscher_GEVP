```python
import matplotlib._pylab_helpers as pylab_helpers

def inspect_all_figures():
    managers = pylab_helpers.Gcf.get_all_fig_managers()

    if not managers:
        print("No live matplotlib figures.")
        return

    for m in managers:
        fig = m.canvas.figure
        print("\n==============================")
        print(f"Figure {fig.number} : '{fig.get_label()}'")
        print("==============================")

        for i, ax in enumerate(fig.axes):
            print(f"\n  Axes {i}")
            print("   title:", ax.get_title())
            print("   xlabel:", ax.get_xlabel())
            print("   ylabel:", ax.get_ylabel())
            print("   xlim:", ax.get_xlim())
            print("   ylim:", ax.get_ylim())
            print("   lines:", len(ax.lines))
            print("   collections:", len(ax.collections))
            print("   patches:", len(ax.patches))
```
